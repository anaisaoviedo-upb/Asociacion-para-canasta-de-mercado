# Reglas de asociación - Canasta de mercado

* Preparación: variables binarias
* Aprendizaje: apriori
* Evaluación: soporte y confianza
* Interpretación: lectura de las reglas

```
# Librerías básicas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Preparación de datos

```
data=pd.read_excel('market.xlsx',sheet_name=0)
data.head()
```

```
#Convertir a datos binarios
data=data.replace({'No': 0, 'Si': 1})
data.head()
```

```
data.info()
```

```
#Corrección del tipo de datos
data=data.astype('category')
data.info()
```

# 2. Aprendizaje - Evaluación



```
from mlxtend.frequent_patterns import association_rules, apriori

#Paso 1 -Productos mas frecuentes: filtrar los que cumplen el soporte
frequent_itemsets = apriori(data, min_support = 0.05, max_len = 2, use_colnames=True)

frequent_itemsets
```

```
# Paso 2 - Armar las reglas verificando confianza
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4)
rules
```

```
rules= rules.sort_values("confidence",ascending=False)
rules
```

```
rules.to_excel('./resultados.xlsx')
```
